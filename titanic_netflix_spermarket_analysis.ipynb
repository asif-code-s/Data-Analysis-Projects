```python
import pandas as pd

df = pd.read_csv('/content/sample_data/3/annex1.csv')

df.head()
```

```python
df.info()
print(df.shape)
```

```python
df.describe(include=['object','number']).transpose()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

itc=df['Item Name'].value_counts()

plt.figure(figsize=(10, 80))  # Wider and taller plot

sns.barplot(x=itc.values,y=itc.index)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a grouped count table
grouped = df.groupby(['Category Name', 'Item Name']).size().reset_index(name='Count')

# Show top 20 to avoid clutter
top = grouped.sort_values('Count', ascending=False).head(20)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(data=top, x='Count', y='Item Name', hue='Category Name')
plt.title("Top Items by Category")
plt.show()
```
